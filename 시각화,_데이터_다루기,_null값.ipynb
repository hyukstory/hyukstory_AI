### 6.2.1 기초 그래프 그리기

```
# seaborn 라이브러리에서 tips 데이터 불러오기

import seaborn as sns

tips = sns.load_dataset('tips')
print(tips.head())
print(type(tips))
```

> #### 1.히스토그램 (hist)
-일변량 그래프

```
fig = plt.figure()
axes1 = fig.add_subplot(1,1,1)

axes1.hist(tips['total_bill'], bins= 20)   # bins : x축의 간격 설정
axes1.set_title('History of Total Bill')
axes1.set_xlabel('Frequency')
axes1.set_ylabel('Total Bill')
```

> #### 2.산점도 (scatter)
-이변량 그래프

```
scatter_plot = plt.figure()
axes1 = scatter_plot.add_subplot(1,1,1)
axes1.scatter(tips['total_bill'], tips['tip'])
axes1.set_title('< Scatterplot of Total Bill vs Tip >')
axes1.set_xlabel('Total Bill')
axes1.set_ylabel('Tip')
```

> #### 3.박스그래프 (boxplot) 

-이산형 변수와 연속형 변수를 함께 사용하는 그래프

```
boxplot = plt.figure()
axes1 = boxplot.add_subplot(1,1,1)

axes1.boxplot([tips[tips['sex'] == 'Female']['tip'],
               tips[tips['sex'] == 'Male']['tip']],
              labels = ['Female', 'Male'])

axes1.set_xlabel('Sex')
axes1.set_ylabel('Tip')
axes1.set_title('Boxplot of Tips by Sex')
```

### 6.2.2 다변량 데이터로 다변량 그래프 그리기 




> 1. 산점도 그래프에 성별 변수 추가

```
# Female, Male 문자열을 정수로 치환하는 함수

def recode_sex(sex) :
  if sex == 'Female' :
    return 0
  else :
    return 1
```

> 2. recode_sex 메서드가 반환한 값(0,1) 을 데이터프레임에 열 새로 생성하여 추가

```
tips['sex_color'] = tips['sex'].apply(recode_sex)
print(tips['sex_color'].head())
```

> 3. 테이블당 인원 수(size)도 산점도 그래프에 추가

```
scatter_plot = plt.figure()
axes1 = scatter_plot.add_subplot(1,1,1)
axes1.scatter(
    x=tips['total_bill'],
    y=tips['tip'],
    s=tips['size']*50 ,    # s는 점의크기
    c=tips['sex_color'],   # c는 점의 색상을 의미
    alpha = 0.5)            # alpha 인자값을 0.5로 지정하여 점의 투명도 조절

axes1.set_title('Total Bill vs Tip Colored by Sex and Sized by Size')
axes1.set_xlabel('Total Bill')
axes1.set_ylabel('Tips')
```

### 6.3.1 단변량 그래프

```
import seaborn as sns
tips = sns.load_dataset('tips')
```

> #### 히스토그램 : subplots, distplot

```
import matplotlib.pyplot as plt

ax = plt.subplots()     # 기본 틀
ax = sns.distplot(tips['total_bill'])   # 열데이터 전달
ax.set_title('Total Bill Histogram with Density Plot')
```

> 밀집도 그래프를 제외하고 싶다면 kde = False 로 설정


> 밀집도 그래프만 나타내려면 hist = False 로 지정

```
ax = plt.subplots()     # 기본 틀
ax = sns.distplot(tips['total_bill'], kde = False , hist = True)  
ax.set_title('Total Bill Histogram with Density Plot')
```

> 양탄자(rug) 그래프 추가 - 그래프의 축에 동일한 길이의 직선을 붙여 데이터의 밀집 정도를 표현

```
ax = plt.subplots()     
ax = sns.distplot(tips['total_bill'], rug = True)   
ax.set_title('Total Bill Histogram with Density Plot')
```

#### count 그래프

```
ax = plt.subplots()     
ax = sns.countplot('day', data = tips)   
ax.set_title('Count of days')
```

### 6.3.2 다변량 그래프


> #### 1.산점도 그래프 : regplot, lmplot

```
# regplot 

ax = plt.subplots()
ax = sns.regplot(x = 'total_bill', y = 'tip',
                 fit_reg = True, # 회귀선 제거하려면 fit_reg = False
                 data = tips)   
ax.set_title('Scatterplot of Total Bill and Tip')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Tip')
```

```
# lmplot

ax = sns.lmplot(x='total_bill', y = 'tip', 
                hue = 'sex',   # 색상 추가하고 싶으면 lmplot (regplot에서는 색상 추가 불가)
                fit_reg=False,   # 회귀선
                markers = ['o', 'x'],  # 마커 설정
                data=tips)

plt.show()
```

> #### 2.joinplot

-산점도 그래프와 히스토그램을 한 번에

```
joint = sns.jointplot(x='total_bill', y='tip', 
                      kind = 'hex',             # 점 -> 육각형
                      data=tips)
joint.set_axis_labels(xlabel='Total Bill', ylabel ='Tip')
joint.fig.suptitle('Joint Plot of Total Bill and Tip', fontsize=10, y= 1.03)
```

> #### 3.kdeplot 이차원 밀집도 

```
ax = plt.subplots()
ax = sns.kdeplot(data = tips['total_bill'],
                 data2=tips['tip'],
                 shade=True)

ax.set_title('Kernel Density Plot of Total Bill and Tip')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Tip')
```

> #### 4.barplot

```
ax = plt.subplots()
ax = sns.barplot(x='time', y = 'total_bill', data=tips)
ax.set_title('Bar plot of average total bill for time of day')
ax.set_xlabel('Time of day')
ax.set_ylabel('Average total bill')
```

> #### 5.boxplot

```
ax = plt.subplots()
ax = sns.boxplot(x='time', y = 'total_bill', data=tips)
ax.set_title('Boxplot of average total bill for time of day')
ax.set_xlabel('Time of day')
ax.set_ylabel('Average total bill')
```

> #### 6.violinplot

```
ax = plt.subplots()
ax = sns.violinplot(x='time', y = 'total_bill',
                    hue='sex',  # 색상 구분할 데이터
                    split= False,  # True : 두 그래프 한번에 / False : 두 그래프 따로 
                    data = tips)
ax.set_title('violinplot of average total bill for time of day')
ax.set_xlabel('Time of day')
ax.set_ylabel('Average total bill')
```

> #### 7.pairplot 관계 그래프

```
graph = sns.pairplot(tips, hue = 'sex')
graph
```

> #### 8.관계 그래프 중복 없애고 지정하기

```
pair_grid = sns.PairGrid(tips)
pair_grid = pair_grid.map_upper(sns.regplot)    # 대각선 위
pair_grid = pair_grid.map_lower(sns.kdeplot,shade=True)    # 대각선 아래
pair_grid = pair_grid.map_diag(sns.boxplot)     # 대각선 
plt.show
```

```
print(row_concat.iloc[3, ])
```

#### 새로운 행 추가

```
# 1개의 행을 가지는 데이터프레임 생성하여 df1에 연결
new_row_df = pd.DataFrame([['n1','n2',' n3','n4']], columns=['A','B','C','D'])
print(new_row_df)

print(df1.append(new_row_df))   #  행이 1개일때 : append

print(pd.concat([df1, new_row_df]))  # 행이 2개 이상일때 : concat
```

ignore_index 인자 사용

```
# 딕셔너리 사용해서 행 추가하기

data_dict = {'A' : 'n1', 'B' : 'n2', 'C' : 'n3', 'D' : 'n4'}
print(df1.append(data_dict, ignore_index = True))   # ignore_index = True : 합치고 나서 첫행부터 다시 인덱스 시작
```

```
row_concat_i = pd.concat([df2, df3], ignore_index = True)
print(row_concat_i)
```

#### 새로운 열 추가

```
# axis 인자 사용
col_concat = pd.concat([df1, df2, df3], axis= 1)   # axis = 0 : 행(기본값),  axis = 1 : 열
print(col_concat)

print(col_concat['A'])


col_concat = pd.concat([df1, df2, df3], axis= 1, ignore_index = True)  
print(col_concat)
```

```
# 더 간단하게 열 추가
col_concat['new_col_list'] = ['n1', 'n2', 'n3', 'n4']
print(col_concat)
```

#### 공통 열과 공통 인덱스만 연결

```
df1.columns = ['A','B','C','D']
df2.columns = ['E','F','G','H']
df3.columns = ['A','C','F','H']

row_concat = pd.concat([df1, df2, df3])
print(row_concat)                       #공통되지 않은 열의 데이터는 누락값으로 처리
```

```
# 공통 열만 추출해서 연결
row_concat = pd.concat([df1, df3], ignore_index = True, join = 'inner')
print(row_concat) 
```

```
# 공통 행만 추출해서 연결  ( axis = 1)
df1.index = [0,1,2,3]
df3.index = [0,2,5,7]

print(df1) ; print(df3)

col_concat = pd.concat([df1, df3], ignore_index = True, axis=1, join = 'inner')
print(col_concat) 
```

# 8.누락값 처리하기

## 8.1 누락값 확인하기

```
ebola = pd.read_csv('/content/drive/MyDrive/dataset/country_timeseries.csv')

print(ebola.isnull())
print(ebola.notnull())
```

```
import numpy as np

print(np.count_nonzero(ebola.isnull()))
print(np.count_nonzero(ebola['Cases_Guinea'].isnull()))
```

## 8.2 누락값 변경하기

누락값을 0으로 변경

```
print(ebola.fillna(0).iloc[0:10, 0:5])
```

바로 위의 값을 받아서 누락값 채우기

```
print(ebola.fillna(method = 'ffill').iloc[0:10, 0:5])
```

바로 다음의 값을 받아서 누락값 채우기

```
print(ebola.fillna(method = 'bfill').iloc[0:10, 0:5])
```

누락값 양쪽에 있는 값 이용하여 중간값으로 처리

```
print(ebola.interpolate().iloc[0:10, 0:5])
```

누락값 삭제하기

```
print(ebola.dropna())
```
